```python
# importing necessary libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# reading files

ds = load_breast_cancer()

# error preventing

low_memory = False    
```

```python
# deleting unnecessary columns that do not provide any crucial information for analysis 

del ds['frame']
del ds['filename']
del ds['data_module']
```

```python
# converting numpy array to a pandas dataframe

df_ds = pd.DataFrame(np.c_[ds['data'], ds['target']], columns = np.append(ds['feature_names'], ['target']))
df_ds.head(5)
```

```python
# MODEL TRAINING: TASK PART 3

x = df_ds.drop(['target'],axis=1)
x
```

```python
y = df_ds['target']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state= 5)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
svc_model = SVC()
svc_model.fit(x_train, y_train)
```

```python
# MODEL EVALUATION: TASK PART 4

y_predict = svc_model.predict(x_test)
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
# MODEL IMPROVEMENT: TASK PART 5

min_train = x_train.min()
min_train
```

```python
range_train = (x_train - min_train).max()
range_train
```

```python
x_train_scaled = (x_train - min_train)/range_train
x_train_scaled
```

```python
sns.scatterplot(x = x_train['mean area'], y= x_train['mean smoothness'], hue= y_train)
```

```python
sns.scatterplot(x= x_train_scaled['mean area'], y= x_train_scaled['mean smoothness'], hue= y_train)
```

```python
min_test = x_test.min()
range_test = (x_test - min_test).max()
x_test_scaled = (x_test - min_test)/ range_test
svc_model.fit(x_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(x_test_scaled)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True, fmt = 'd')
```

```python
print(classification_report(y_test, y_predict))
```
